```python
# Three-Membered Community GCM Reconstruction Script

# For This Study We have Considered these abbreviations for GCMs Construction
# Please Change the abbreviations according to your models

# pr: (Prevotella ruminicola/Xylanibacter ruminicola)
# bf: (Butyrivibrio fibrisolvens)
# sr: (Selenomonas ruminantium)
# br: (Bifidobacterium ruminantium)
# lr: (Ligilactobacilus ruminis)
# ra: (Ruminococcus albus)

import cobra
import sys
import os

model1 = cobra.io.read_sbml_model('Model_1')
model1.slim_optimize()

model2 = cobra.io.read_sbml_model('Model_2')
model2.slim_optimize()

model3 = cobra.io.read_sbml_model('Model_3')
model3.slim_optimize()

print(len(model1.reactions))
print(len(model1.metabolites))
print(len(model1.genes))

solution1=model1.optimize()
solution1.objective_value

print(len(model2.reactions))
print(len(model2.metabolites))
print(len(model2.genes))

solution2=model2.optimize()
solution2.objective_value

print(len(model3.reactions))
print(len(model3.metabolites))
print(len(model3.genes))

solution3=model3.optimize()
solution3.objective_value

for i in model1.metabolites:
    if(str(i)[-2]=="c"):
        j=str(i)+"bf"        # Considered As Model_1 (Change Accordingly)
        i.id=j
        i.compartment="c0bf" # Considered As Model_1 (Change Accordingly)
        
for i in model1.metabolites:
    if(str(i)[-2]=="c"):
        j=str(i)+"bf"        # Considered As Model_2 (Change Accordingly)
        i.id=j
        i.compartment="c0bf" # Considered As Model_2 (Change Accordingly)
        
for i in model3.metabolites:
    if(str(i)[-2]=="c"):
        j=str(i)+"lr"        # Considered As Model_3 (Change Accordingly)
        i.id=j
        i.compartment="c0lr" # Considered As Model_3 (Change Accordingly)
        
count=0
for i in model1.reactions:
    if(str(i.id)[0]=="r"):
        j=str(i.id)+"bf"     # Considered As Model_1 (Change Accordingly)
        i.id=j
        count=count+1
print(count)

count=0
for i in model2.reactions:
    if(str(i.id)[0]=="r"):
        j=str(i.id)+"br"     # Considered As Model_2 (Change Accordingly)
        i.id=j
        count=count+1
print(count)

count=0
for i in model3.reactions:
    if(str(i.id)[0]=="r"):
        j=str(i.id)+"lr"     # Considered As Model_3 (Change Accordingly)
        i.id=j
        count=count+1
print(count)

count=0
for i in model1.reactions:
    if(str(i.id)=="bio1"):
        j=str(i.id)+"bf"
        i.id=j
        count=count+1
print(count)

count=0
for i in model2.reactions:
    if(str(i.id)=="bio1"):
        j=str(i.id)+"br"
        i.id=j
        count=count+1
print(count)

count=0
for i in model3.reactions:
    if(str(i.id)=="bio1"):
        j=str(i.id)+"lr"
        i.id=j
        count=count+1
print(count)

count=0
for i in model1.reactions:
    if(str(i.id)=="EX_cpd11416_c0"):continue
    if(str(i.id)[0]=="E" and str(i.id)[-2]=="c"):
        j=str(i.id)+"bf"
        i.id=j
        count=count+1
print(count)

count=0
for i in model2.reactions:
    if(str(i.id)=="EX_cpd11416_c0"):continue
    if(str(i.id)[0]=="E" and str(i.id)[-2]=="c"):
        j=str(i.id)+"br"
        i.id=j
        count=count+1
print(count)

count=0
for i in model3.reactions:
    if(str(i.id)=="EX_cpd11416_c0"):continue
    if(str(i.id)[0]=="E" and str(i.id)[-2]=="c"):
        j=str(i.id)+"lr"
        i.id=j
        count=count+1
print(count)

from cobra import Model, Reaction, Metabolite
model = Model('Gut_community_model')

count=0
for i in model1.reactions:
    if(str(i.id)=="EX_cpd11416_c0"):
        continue
    reaction=model1.reactions.get_by_id(str(i.id))
    reaction.reaction
    count=count+1
    model.add_reactions([reaction])
        
print(count)

count=0
for i in model2.reactions:
    if(str(i.id)=="EX_cpd11416_c0"):
        continue
    reaction=model2.reactions.get_by_id(str(i.id))
    reaction.reaction
    count=count+1
    model.add_reactions([reaction])
        
print(count)

count=0
for i in model3.reactions:
    if(str(i.id)=="EX_cpd11416_c0"):
        continue
    reaction=model3.reactions.get_by_id(str(i.id))
    reaction.reaction
    count=count+1
    model.add_reactions([reaction])
        
print(count)

print(len(model.reactions))
print(len(model.metabolites))
print(len(model.genes))

from cobra import Metabolite
cpd11416_c0 = Metabolite(
    'cpd11416_c0',
    name='Complete_Biomass',
    compartment='c0'
    )

from cobra import Metabolite
cpd11416_b = Metabolite(
    'cpd11416_b',
    name='Complete_Biomass',
    compartment='e0'
    )

reaction = Reaction('Bio1')
reaction.name = ('Community_biomass')
reaction.lower_bound = 0
reaction.upper_bound = 1000

reaction.add_metabolites({model1.metabolites.get_by_id("cpd11416_c0bf"): -0.333333, # For Model_1
                          model2.metabolites.get_by_id("cpd11416_c0br"): -0.333333, # For Model_2
                          model3.metabolites.get_by_id("cpd11416_c0lr"): -0.333333, # For Model_3
                          cpd11416_c0: 1
                         })

        
reaction.reaction
model.add_reactions([reaction])
print(count)

reaction = Reaction('EX_cpd11416_c0')
reaction.name = ('EX_cpd11416_c0')
reaction.lower_bound = 0
reaction.upper_bound = 1000

reaction.add_metabolites({
                          model.metabolites.get_by_id("cpd11416_c0"): -1,
                          cpd11416_b: 1
                        })

        
reaction.reaction
model.add_reactions([reaction])

print(len(model.reactions))
print(len(model.metabolites))
print(len(model.genes))

model.objective = 'Bio1'

model.slim_optimize()

## If GCM Model do not show any Growth rate...
## Check the IndividuL GEM-Specific Media Conditions and Their Bounds of EXchange Recations
## For this case the bounds of the below reaction was (0.0, 0.0) after GCM reconstruction

model.reactions.EX_cpd00305_b.bounds = (-1000.0, 1000.0)

print(model.summary())

cobra.io.write_sbml_model(model,'Three-Membered/BF_BR_LR.sbml')
```
