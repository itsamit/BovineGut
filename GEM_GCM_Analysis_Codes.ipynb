```python
## Biomass Sensitivity Analysis (For Individual GEM Only) ##

## STEP 1 ## (Sensitivity Analysis)

import cobra
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns
%matplotlib inline


## STEP 2 ## (Sensitivity Analysis)

print(len(model.reactions.bio1.reactants))

## STEP 3 ## (Sensitivity Analysis)

bio=model.reactions.get_by_id('bio1')
met_pc=bio.metabolites
met_names=[]
met_coeff=[]
met_formula=[]
met_chemnames=[]
for met in met_pc:
    met_names.append(met.id)
    met_coeff.append(met_pc[met])
    
## STEP 4 ## (Sensitivity Analysis)

def gr(p, metc, metn):
    
    # If not biomass precursor (i.e., positive coefficient), skip safely
    if metc >= 0:
        return np.nan, np.nan

    try:
        # -------- Increase --------
        modelF = model.copy()
        biof = modelF.reactions.get_by_id('bio1')
        
        biof.add_metabolites({
            modelF.metabolites.get_by_id(metn): -p * metc
        })
        
        sol_plus = modelF.slim_optimize()

        # -------- Decrease --------
        modelF = model.copy()
        biof = modelF.reactions.get_by_id('bio1')
        
        biof.add_metabolites({
            modelF.metabolites.get_by_id(metn): p * metc
        })
        
        sol_minus = modelF.slim_optimize()

        return sol_plus, sol_minus

    except Exception as e:
        print(f"Error in {metn}: {e}")
        return np.nan, np.nan
    
    
## STEP 5 ## (Sensitivity Analysis)

p = 15
p = p/100
p10 =[]
n10 =[]
for metn,metc in zip(met_names,met_coeff):
    print('Current Biomass Precursor: ',metn)
    grp10, grn10 = gr(p,metc,metn)
    p10.append(grp10)
    n10.append(grn10)
    print('Increased growth rate: ', grp10)
    print('Decrease growth rate: ', grn10)
    

## STEP 6 ## (Sensitivity Analysis)

import pandas as pd

df = pd.DataFrame({
    "Metabolite": met_names,
    "Coeff": met_coeff,
    "Growth_Increase": p10,
    "Growth_Decrease": n10
})

df.head()
# print(df)


## STEP 7 ## (Save Summary of the Model Sensitivity Analysis)

df.to_excel("Model_Sensitivity.xlsx")

## STEP 8 ## (Sort Sensitivity Analysis Data For Plotting)

df_sorted=pd.DataFrame({"Name":pd.Series(met_names),"Increase":p10,"Decrease":n10,'Coeff':met_coeff})
df_sorted['Diff']=df_sorted['Increase']-df_sorted['Decrease']
df_sorted=df_sorted.sort_values(by='Diff',ascending=False)
df_sorted=df_sorted.reset_index(drop=True)

# STEP 9 ## (Save Sensitivity Analysis Figure)

import matplotlib.ticker as ticker
fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(21,10))
ax.bar(df_sorted['Name'],df_sorted['Diff'],bottom=df_sorted['Decrease'],color='gray')
ax.set_ylabel('Growth rate (1/hr)',fontsize=12)
ax.set_xlabel('Biomass Precursors',fontsize=12)
ax.set_ylim([0.685,0.725]) # Set Axis Lentgh Based on Analysis Data
#ax.set_xlim([1,55])
#ax.set_yticks(df_sorted['Increase'])
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
#ax.set_xticklabels('A','B','C')
ax.plot(df_sorted['Increase'],'go')
ax.plot(df_sorted['Decrease'],'ro')
ax.legend(['15% increase in coefficients','15% decrease in coefficients'],fontsize=15)
plt.gca()
plt.savefig('Model_Sensitivity.jpeg',dpi=600)
```

```python
## Biomass Sensitivity Analysis (For Community GCM Only) ##

## STEP 1 ## (Community-Based Sensitivity Analysis (Bio1))

import cobra
import numpy as np
import pandas as pd

# ---- load model ----
model = cobra.io.read_sbml_model("Model") # (Add Model Directory)

# ---- reaction/metabolite IDs based on SBML Model----
bio_rxn_id = "Bio1"                 # community biomass reaction
A_met_id = "cpd11416_c0sr"        # Species_1 biomass metabolite
B_met_id = "cpd11416_c0lr"        # Species_2 / second member biomass metabolite
comm_met_id = "cpd11416_c0"        # community biomass metabolite

bio_rxn = model.reactions.get_by_id(bio_rxn_id)
A_met = model.metabolites.get_by_id(A_met_id)
B_met = model.metabolites.get_by_id(B_met_id)
comm_met = model.metabolites.get_by_id(comm_met_id)

# optional: make sure objective is Bio1
model.objective = bio_rxn

def set_biomass_coefficients(model, bio_rxn_id, A_met_id, B_met_id, comm_met_id, A_coeff):
    """
    Set community biomass reaction to:
    A_coeff * A + (1 - A_coeff) * B --> 1 * community_biomass
    """
    B_coeff = 1.0 - A_coeff

    rxn = model.reactions.get_by_id(bio_rxn_id)
    A = model.metabolites.get_by_id(A_met_id)
    B = model.metabolites.get_by_id(B_met_id)
    comm = model.metabolites.get_by_id(comm_met_id)

    # current coefficients
    current_A = rxn.get_coefficient(A.id)
    current_B = rxn.get_coefficient(B.id)
    current_comm = rxn.get_coefficient(comm.id)

    # remove old values and set new ones
    rxn.add_metabolites({
        A: -A_coeff - current_A,
        B: -B_coeff - current_B,
        comm: 1.0 - current_comm
    }, combine=True)

def scan_biomass_coefficients(model, start=0.0, stop=1.0, step=0.05):
    results = []

    coeffs = np.arange(start, stop + step/2, step)

    for A_coeff in coeffs:
        with model:
            set_biomass_coefficients(
                model=model,
                bio_rxn_id=bio_rxn_id,
                A_met_id=A_met_id,
                B_met_id=B_met_id,
                comm_met_id=comm_met_id,
                A_coeff=A_coeff
            )

            sol = model.optimize()

            results.append({
                "A_coeff": A_coeff,
                "B_coeff": 1.0 - A_coeff,
                "Bio1_flux": sol.objective_value,
                "status": sol.status
            })

    return pd.DataFrame(results)

# ---- run scan ----
df = scan_biomass_coefficients(model, start=0.0, stop=1.0, step=0.1)

# ---- save results ----
df.to_csv("biomass_coefficient_scan.csv", index=False)

print(df)
```

```python
"""=============================================================================
Compartmentalized Flux Distribution Pathway Map
Full-model visualization with compartment separation
=============================================================================

This script generates a complete pathway map showing:
  - All active reactions organized by metabolic compartment
  - Compartment-to-compartment flux transfers (transport reactions)
  - Pathway distribution within each compartment
  - Metabolite hub analysis per compartment
  - Inter-compartment connectivity

WHAT THIS DOES
--------------
1. Analyzes all compartments in the model (cytosol, extracellular, etc.)
2. Maps reactions to compartments based on their metabolites
3. Identifies transport reactions (cross-compartment)
4. Creates compartmentalized pathway maps
5. Generates multi-panel visualizations showing:
     - Compartment-level flux summary
     - Pathway distribution per compartment
     - Transport flux between compartments
     - Metabolite hubs in each compartment

REQUIREMENTS
------------
  pip install cobra pandas numpy matplotlib networkx

USAGE
-----
  python compartmentalized_pathway_map.py

OUTPUT FILES
------------
  compartment_summary.csv              - flux summary per compartment
  compartment_pathways.csv             - pathway distribution per compartment
  transport_reactions_2.csv            - inter-compartment flux
  compartment_metabolites.csv          - metabolite hubs per compartment
  compartment_flux_map.png/svg/pdf     - visual pathway map
  compartment_network.graphml          - full network with compartments

# =============================================================================
# """

import cobra
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import FancyBboxPatch, Rectangle, FancyArrowPatch
import json
import warnings
warnings.filterwarnings("ignore")

# Configuration
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans']
plt.rcParams['font.size'] = 9
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['savefig.bbox'] = 'tight'

# File paths
MODEL_FILE  = "Model" # Paste the Model Path #
BIOMASS_RXN = "bio1"
FLUX_CUTOFF = 1e-6

# Color schemes
COMPARTMENT_COLORS = {
    'c': '#58a6ff',   # cytosol - blue
    'c0': '#58a6ff',  # cytosol variant
    'e': '#3fb950',   # extracellular - green
    'e0': '#3fb950',  # extracellular variant (Add compartments Based on GEMs/GCMs)
}

PATHWAY_COLORS = [
    "#1f77b4",  # blue
    "#ff7f0e",  # orange
    "#2ca02c",  # green
    "#d62728",  # red
    "#9467bd",  # purple
    "#8c564b",  # brown
    "#e377c2",  # pink
    "#7f7f7f",  # gray
    "#bcbd22",  # olive
    "#17becf",  # cyan
    "#393b79",  # dark blue
    "#637939",  # dark olive
    "#8c6d31"   # mustard brown
]


# =============================================================================
# 1. LOAD MODEL AND RUN FBA
# =============================================================================

print("=" * 70)
print("COMPARTMENTALIZED FLUX DISTRIBUTION PATHWAY MAP")
print("=" * 70)
print("\nLoading model and running FBA...")

try:
    model = cobra.io.read_sbml_model(MODEL_FILE)
    print(f"  ✓ Model loaded: {len(model.reactions)} reactions, "
          f"{len(model.metabolites)} metabolites")
except Exception as e:
    print(f"[ERROR] Failed to load model: {e}")
    exit(1)

# Run FBA
try:
    solution = model.optimize()
    if solution.status != "optimal":
        print(f"[ERROR] FBA failed with status: {solution.status}")
        exit(1)
    
    flux_series = solution.fluxes
    growth = solution.objective_value
    print(f"  ✓ FBA complete: growth = {growth:.6f} mmol/gDW/h")
    print(f"  ✓ Active reactions: {(flux_series.abs() > FLUX_CUTOFF).sum()}")
except Exception as e:
    print(f"[ERROR] FBA optimization failed: {e}")
    exit(1)


# =============================================================================
# 2. ANALYZE COMPARTMENTS
# =============================================================================

print("\n" + "=" * 70)
print("Analyzing compartments...")

# Extract all compartments from metabolites
compartments = {}
for met in model.metabolites:
    comp_id = met.compartment
    if comp_id not in compartments:
        # Try to get compartment name from model
        comp_name = model.compartments.get(comp_id, comp_id)
        compartments[comp_id] = {
            'id': comp_id,
            'name': comp_name,
            'metabolites': [],
            'reactions': set(),
            'total_flux': 0.0,
        }
    compartments[comp_id]['metabolites'].append(met)

print(f"\n  Found {len(compartments)} compartments:")
for comp_id, comp_data in sorted(compartments.items()):
    print(f"    {comp_id:10} - {comp_data['name']:30} "
          f"({len(comp_data['metabolites'])} metabolites)")


# =============================================================================
# 3. CLASSIFY REACTIONS BY COMPARTMENT
# =============================================================================

print("\n" + "=" * 70)
print("Classifying reactions by compartment...")

PATHWAY_KEYWORDS = {
    "Glycolysis / EMP"           : ["glycolysis","glucose","fructose","phosphoglycerate",
                                    "pyruvate kinase","phosphofructo","aldolase","enolase",
                                    "gapdh","pgk","pgm","pgi","pfk","tpi","pyk","eno"],
    "Pentose Phosphate Pathway"  : ["pentose","ribose","ribulose","xylulose","transketolase",
                                    "transaldolase","glucose-6-phosphate dehydrogenase",
                                    "6-phosphogluconate","sedoheptulose"],
    "TCA Cycle"                  : ["citrate","isocitrate","aconitate","succinate","fumarate",
                                    "malate","oxaloacetate","alpha-ketoglutarate","2-oxoglutarate",
                                    "succinyl","tca","krebs","citric acid"],
    "Pyruvate Metabolism"        : ["pyruvate","acetyl-coa","acetaldehyde","acetate","formate",
                                    "lactate","pyruvate decarboxylase","pyruvate ferredoxin",
                                    "pyruvate formate","pfl","pdh","pdc"],
    "Fatty Acid Metabolism"      : ["fatty acid","acyl","malonyl","beta-oxidation","enoyl",
                                    "ketoacyl","butyryl","crotonyl","fas","fab"],
    "Amino Acid Metabolism"      : ["glutamate","glutamine","alanine","aspartate","serine",
                                    "threonine","leucine","valine","isoleucine","arginine",
                                    "proline","lysine","methionine","cysteine","tryptophan",
                                    "phenylalanine","tyrosine","histidine","asparagine"],
    "Nucleotide Metabolism"      : ["purine","pyrimidine","adenine","guanine","cytosine",
                                    "thymine","uracil","inosine","xanthine","amp","gmp",
                                    "cmp","ump","atp synthase"],
    "Cell Wall / Peptidoglycan"  : ["peptidoglycan","murein","mur","cell wall","glucosamine",
                                    "n-acetyl","udp-glc","udp-gal"],
    "Cellulose Degradation"      : ["cellulose","cellobiose","cellodextrin","beta-glucosidase",
                                    "endoglucanase","exoglucanase","xylan","xylose",
                                    "arabinose","hemicellulose"],
    "Acetogenesis / Fermentation": ["acetate kinase","phosphotransacetylase","acetyl-coa",
                                    "ethanol","butanol","butyrate","acetoin","pta","ack"],
    "CO2 / One-Carbon"           : ["co2","carbon dioxide","formate","methyl","thf",
                                    "tetrahydrofolate","methylene","methenyl","formyl"],
    "Transport"                  : ["transport","exchange","uptake","export","import",
                                    "permease","abc transporter","symport","antiport"],
    "Oxidative Phosphorylation"  : ["nadh dehydrogenase","complex i","complex ii","complex iii",
                                    "complex iv","atp synthase","electron transport",
                                    "cytochrome","quinone","menaquinone"],
}

def classify_pathway(rxn_id, rxn_name, subsystem):
    if subsystem and subsystem.strip() and subsystem.strip().lower() != "unknown":
        return subsystem.strip()
    text = (rxn_id + " " + rxn_name).lower()
    for pathway, keywords in PATHWAY_KEYWORDS.items():
        if any(kw in text for kw in keywords):
            return pathway
    return "Other / Unclassified"

# Analyze each reaction
reaction_data = []
transport_reactions = []

for rxn in model.reactions:
    flux = flux_series.get(rxn.id, 0.0)
    if abs(flux) < FLUX_CUTOFF:
        continue
    
    # Get compartments of all metabolites in this reaction
    rxn_compartments = set(met.compartment for met in rxn.metabolites.keys())
    
    # Classify as transport if crosses compartments
    is_transport = len(rxn_compartments) > 1
    
    # Primary compartment (most metabolites)
    comp_counts = {}
    for met in rxn.metabolites.keys():
        comp = met.compartment
        comp_counts[comp] = comp_counts.get(comp, 0) + 1
    primary_comp = max(comp_counts.items(), key=lambda x: x[1])[0] if comp_counts else 'unknown'
    
    # Pathway classification
    pathway = classify_pathway(rxn.id, rxn.name or "", rxn.subsystem or "")
    
    rxn_record = {
        'reaction_id': rxn.id,
        'reaction_name': rxn.name or rxn.id,
        'flux': round(flux, 6),
        'abs_flux': abs(flux),
        'direction': 'forward' if flux > 0 else 'reverse',
        'primary_compartment': primary_comp,
        'compartments': ';'.join(sorted(rxn_compartments)),
        'n_compartments': len(rxn_compartments),
        'is_transport': is_transport,
        'pathway': pathway,
        'subsystem': rxn.subsystem or "Unknown",
        'equation': rxn.reaction,
    }
    
    reaction_data.append(rxn_record)
    
    # Track compartment flux
    if primary_comp in compartments:
        compartments[primary_comp]['reactions'].add(rxn.id)
        compartments[primary_comp]['total_flux'] += abs(flux)
    
    # Transport reactions
    if is_transport:
        for comp_from in rxn_compartments:
            for comp_to in rxn_compartments:
                if comp_from != comp_to:
                    transport_reactions.append({
                        'reaction_id': rxn.id,
                        'reaction_name': rxn.name or rxn.id,
                        'from_compartment': comp_from,
                        'to_compartment': comp_to,
                        'flux': abs(flux),
                        'pathway': pathway,
                    })

reaction_df = pd.DataFrame(reaction_data)
transport_df = pd.DataFrame(transport_reactions)

print(f"\n  ✓ Total active reactions: {len(reaction_df)}")
print(f"  ✓ Transport reactions: {len(reaction_df[reaction_df['is_transport']])}")
print(f"  ✓ Unique transport routes: {len(transport_df)}")


# =============================================================================
# 4. COMPARTMENT SUMMARY
# =============================================================================

print("\n" + "=" * 70)
print("Compartment-level flux summary...")

comp_summary = []
for comp_id, comp_data in compartments.items():
    n_rxns = len(comp_data['reactions'])
    if n_rxns == 0:
        continue
    
    comp_rxns = reaction_df[reaction_df['primary_compartment'] == comp_id]
    
    comp_summary.append({
        'compartment_id': comp_id,
        'compartment_name': comp_data['name'],
        'n_metabolites': len(comp_data['metabolites']),
        'n_reactions': n_rxns,
        'total_flux': comp_data['total_flux'],
        'max_flux': comp_rxns['abs_flux'].max() if len(comp_rxns) > 0 else 0,
        'mean_flux': comp_rxns['abs_flux'].mean() if len(comp_rxns) > 0 else 0,
    })

comp_summary_df = pd.DataFrame(comp_summary).sort_values('total_flux', ascending=False)
comp_summary_df['flux_fraction_%'] = (
    comp_summary_df['total_flux'] / comp_summary_df['total_flux'].sum() * 100
).round(2)

print(f"\n  {'Compartment':<15} {'Name':<25} {'Rxns':>6} {'Mets':>6} {'Total Flux':>12} {'%':>6}")
print("  " + "-" * 80)
for _, row in comp_summary_df.iterrows():
    print(f"  {row['compartment_id']:<15} {row['compartment_name']:<25} "
          f"{row['n_reactions']:>6} {row['n_metabolites']:>6} "
          f"{row['total_flux']:>12.2f} {row['flux_fraction_%']:>6.1f}%")

comp_summary_df.to_csv("compartment_summary.csv", index=False)
print("\n  ✓ Saved → compartment_summary.csv")


# =============================================================================
# 5. PATHWAY DISTRIBUTION PER COMPARTMENT
# =============================================================================

print("\n" + "=" * 70)
print("Analyzing pathway distribution per compartment...")

comp_pathway_data = []
for comp_id in compartments.keys():
    comp_rxns = reaction_df[reaction_df['primary_compartment'] == comp_id]
    if len(comp_rxns) == 0:
        continue
    
    pathway_summary = (
        comp_rxns.groupby('pathway')
        .agg(
            n_reactions=('reaction_id', 'count'),
            total_flux=('abs_flux', 'sum'),
        )
        .reset_index()
    )
    
    for _, row in pathway_summary.iterrows():
        comp_pathway_data.append({
            'compartment_id': comp_id,
            'compartment_name': compartments[comp_id]['name'],
            'pathway': row['pathway'],
            'n_reactions': row['n_reactions'],
            'total_flux': row['total_flux'],
        })

comp_pathway_df = pd.DataFrame(comp_pathway_data)
comp_pathway_df.to_csv("compartment_pathways.csv", index=False)
print("  ✓ Saved → compartment_pathways.csv")

# Transport summary
if len(transport_df) > 0:
    transport_summary = (
        transport_df.groupby(['from_compartment', 'to_compartment'])
        .agg(
            n_reactions=('reaction_id', 'count'),
            total_flux=('flux', 'sum'),
        )
        .reset_index()
        .sort_values('total_flux', ascending=False)
    )
    transport_summary.to_csv("transport_reactions_2.csv", index=False)
    print("  ✓ Saved → transport_reactions.csv")


# =============================================================================
# 6. METABOLITE HUBS PER COMPARTMENT
# =============================================================================

print("\n" + "=" * 70)
print("Identifying metabolite hubs per compartment...")

met_hub_data = []
for comp_id, comp_data in compartments.items():
    comp_rxns_ids = comp_data['reactions']
    if len(comp_rxns_ids) == 0:
        continue
    
    # Analyze metabolite connectivity in this compartment
    met_flux = {}
    for met in comp_data['metabolites']:
        in_flux = 0.0
        out_flux = 0.0
        
        for rxn in met.reactions:
            if rxn.id not in comp_rxns_ids:
                continue
            flux = flux_series.get(rxn.id, 0.0)
            if abs(flux) < FLUX_CUTOFF:
                continue
            
            stoich = rxn.metabolites[met]
            net_stoich = stoich * (1 if flux > 0 else -1)
            
            if net_stoich < 0:
                out_flux += abs(flux * stoich)
            else:
                in_flux += abs(flux * stoich)
        
        total = in_flux + out_flux
        if total > FLUX_CUTOFF:
            met_hub_data.append({
                'compartment_id': comp_id,
                'compartment_name': comp_data['name'],
                'metabolite_id': met.id,
                'metabolite_name': met.name or met.id,
                'formula': met.formula or "",
                'in_flux': round(in_flux, 4),
                'out_flux': round(out_flux, 4),
                'total_flux': round(total, 4),
            })

met_hub_df = pd.DataFrame(met_hub_data).sort_values('total_flux', ascending=False)
met_hub_df.to_csv("compartment_metabolites.csv", index=False)
print(f"  ✓ Identified {len(met_hub_df)} active metabolite hubs")
print("  ✓ Saved → compartment_metabolites.csv")


# =============================================================================
# 7. NETWORK EXPORT WITH COMPARTMENT INFO (Not required)
# =============================================================================

print("\n" + "=" * 70)
print("Exporting compartmentalized network...")

try:
    import networkx as nx
    
    G = nx.DiGraph()
    
    # Add metabolite nodes with compartment info
    for _, row in met_hub_df.iterrows():
        G.add_node(row['metabolite_id'],
                   label=row['metabolite_name'],
                   node_type='metabolite',
                   compartment=row['compartment_id'],
                   compartment_name=row['compartment_name'],
                   formula=row['formula'],
                   total_flux=float(row['total_flux']))
    
    # Add reaction nodes
    for _, row in reaction_df.iterrows():
        G.add_node(row['reaction_id'],
                   label=row['reaction_name'],
                   node_type='reaction',
                   compartment=row['primary_compartment'],
                   is_transport=row['is_transport'],
                   flux=float(row['flux']),
                   pathway=row['pathway'])
    
    # Add edges
    for rxn in model.reactions:
        flux = flux_series.get(rxn.id, 0.0)
        if abs(flux) < FLUX_CUTOFF:
            continue
        
        for met, stoich in rxn.metabolites.items():
            net_stoich = stoich * (1 if flux > 0 else -1)
            if net_stoich < 0:
                G.add_edge(met.id, rxn.id, flux=abs(flux), stoich=abs(stoich))
            else:
                G.add_edge(rxn.id, met.id, flux=abs(flux), stoich=abs(stoich))
    
    nx.write_graphml(G, "compartment_network.graphml")
    print("  ✓ Saved → compartment_network.graphml")
    
except ImportError:
    print("  [SKIP] networkx not installed")


# =============================================================================
# 8. COMPARTMENTALIZED FLUX MAP VISUALIZATION
# =============================================================================

print("\n" + "=" * 70)
print("Creating compartmentalized flux map visualization...")

fig = plt.figure(figsize=(20, 14))
fig.patch.set_facecolor('#0d1117')

gs = fig.add_gridspec(3, 3, hspace=0.35, wspace=0.35,
                      left=0.05, right=0.95, top=0.92, bottom=0.05)

# ── Panel A: Compartment flux summary (bar chart) ──
ax1 = fig.add_subplot(gs[0, :2])
ax1.set_facecolor('#161b22')

y_pos = np.arange(len(comp_summary_df))
colors = [COMPARTMENT_COLORS.get(c, '#8b949e') for c in comp_summary_df['compartment_id']]

ax1.barh(y_pos, comp_summary_df['total_flux'], color=colors,
         edgecolor='#30363d', linewidth=1.5, height=0.7)

for i, (_, row) in enumerate(comp_summary_df.iterrows()):
    ax1.text(row['total_flux'] + max(comp_summary_df['total_flux']) * 0.02, i,
            f"{row['total_flux']:.1f}  ({row['flux_fraction_%']:.1f}%)",
            va='center', fontsize=9, color='#e6edf3', fontweight='bold')

ax1.set_yticks(y_pos)
labels = [f"{row['compartment_id']} - {row['compartment_name']}" 
          for _, row in comp_summary_df.iterrows()]
ax1.set_yticklabels(labels, fontsize=9, color='#e6edf3', fontweight='bold')
ax1.set_xlabel('Total Flux (mmol/gDW/h)', color='#e6edf3', fontsize=10, fontweight='bold')
ax1.set_title('A. Flux Distribution by Compartment', color='#58a6ff',
             fontsize=12, fontweight='bold', loc='left')
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['bottom'].set_color('#30363d')
ax1.spines['left'].set_color('#30363d')
ax1.tick_params(colors='#8b949e')
ax1.grid(axis='x', alpha=0.2, color='#30363d')

# ── Panel B: Transport flux matrix (heatmap) ──
ax2 = fig.add_subplot(gs[0, 2])
ax2.set_facecolor('#161b22')

if len(transport_df) > 0 and 'from_compartment' in transport_summary.columns:
    # Create flux matrix
    all_comps = sorted(set(transport_summary['from_compartment']) | 
                      set(transport_summary['to_compartment']))
    n = len(all_comps)
    flux_matrix = np.zeros((n, n))
    
    for _, row in transport_summary.iterrows():
        i = all_comps.index(row['from_compartment'])
        j = all_comps.index(row['to_compartment'])
        flux_matrix[i, j] = row['total_flux']
    
    im = ax2.imshow(flux_matrix, cmap='YlOrRd', aspect='auto')
    ax2.set_xticks(range(n))
    ax2.set_yticks(range(n))
    ax2.set_xticklabels(all_comps, fontsize=8, color='#e6edf3', rotation=45, ha='right')
    ax2.set_yticklabels(all_comps, fontsize=8, color='#e6edf3')
    ax2.set_xlabel('To Compartment', color='#e6edf3', fontsize=9)
    ax2.set_ylabel('From Compartment', color='#e6edf3', fontsize=9)
    ax2.set_title('B. Transport Flux', color='#58a6ff', fontsize=12, 
                 fontweight='bold', loc='left')
    
    # Colorbar
    cbar = plt.colorbar(im, ax=ax2, fraction=0.046, pad=0.04)
    cbar.set_label('Flux', color='#e6edf3', fontsize=8)
    cbar.ax.tick_params(colors='#8b949e', labelsize=7)
else:
    ax2.text(0.5, 0.5, 'No transport\nreactions detected',
            ha='center', va='center', fontsize=10, color='#8b949e')
    ax2.axis('off')

# ── Panel C: Top pathways per compartment (stacked bars) ──
ax3 = fig.add_subplot(gs[1, :])
ax3.set_facecolor('#161b22')

# Prepare data for stacked bar
top_comps = comp_summary_df.head(6)['compartment_id'].tolist()
pathway_list = comp_pathway_df['pathway'].unique()[:10]

data_matrix = []
for comp in top_comps:
    comp_data = comp_pathway_df[comp_pathway_df['compartment_id'] == comp]
    row = [comp_data[comp_data['pathway'] == pw]['total_flux'].sum() 
           for pw in pathway_list]
    data_matrix.append(row)

data_matrix = np.array(data_matrix).T

x = np.arange(len(top_comps))
bottom = np.zeros(len(top_comps))

for i, pathway in enumerate(pathway_list):
    color = PATHWAY_COLORS[i % len(PATHWAY_COLORS)]
    ax3.bar(x, data_matrix[i], bottom=bottom, color=color, 
           edgecolor='#30363d', linewidth=0.5, label=pathway[:30])
    bottom += data_matrix[i]

ax3.set_xticks(x)
ax3.set_xticklabels([compartments[c]['name'][:20] for c in top_comps],
                     fontsize=9, color='#e6edf3', rotation=15, ha='right')
ax3.set_ylabel('Flux (mmol/gDW/h)', color='#e6edf3', fontsize=10)
ax3.set_title('C. Pathway Distribution per Compartment (Stacked)',
             color='#58a6ff', fontsize=12, fontweight='bold', loc='left')
ax3.spines['top'].set_visible(False)
ax3.spines['right'].set_visible(False)
ax3.spines['bottom'].set_color('#30363d')
ax3.spines['left'].set_color('#30363d')
ax3.tick_params(colors='#8b949e')
ax3.legend(fontsize=7, loc='upper right', facecolor='#161b22', 
          edgecolor='#30363d', labelcolor='#e6edf3', ncol=2)
ax3.grid(axis='y', alpha=0.2, color='#30363d')

# ── Panel D: Top metabolite hubs per compartment ──
ax4 = fig.add_subplot(gs[2, :])
ax4.set_facecolor('#161b22')

# Show top 5 metabolites from top 3 compartments
plot_data = []
for comp in top_comps[:3]:
    comp_mets = met_hub_df[met_hub_df['compartment_id'] == comp].head(5)
    for _, row in comp_mets.iterrows():
        plot_data.append({
            'label': f"{row['metabolite_name'][:30]} ({comp})",
            'flux': row['total_flux'],
            'compartment': comp,
        })

if plot_data:
    plot_df = pd.DataFrame(plot_data)
    y_pos = np.arange(len(plot_df))
    colors = [COMPARTMENT_COLORS.get(c, '#8b949e') for c in plot_df['compartment']]
    
    ax4.barh(y_pos, plot_df['flux'], color=colors, edgecolor='#30363d',
            linewidth=1, height=0.7)
    ax4.set_yticks(y_pos)
    ax4.set_yticklabels(plot_df['label'], fontsize=7, color='#e6edf3')
    ax4.set_xlabel('Total Flux (mmol/gDW/h)', color='#e6edf3', fontsize=10)
    ax4.set_title('D. Top Metabolite Hubs by Compartment',
                 color='#58a6ff', fontsize=12, fontweight='bold', loc='left')
    ax4.spines['top'].set_visible(False)
    ax4.spines['right'].set_visible(False)
    ax4.spines['bottom'].set_color('#30363d')
    ax4.spines['left'].set_color('#30363d')
    ax4.tick_params(colors='#8b949e')
    ax4.grid(axis='x', alpha=0.2, color='#30363d')

# Main title
fig.suptitle(f'Compartmentalized Flux Distribution — Model\n'
            f'Growth: {growth:.6f} mmol/gDW/h  |  {len(compartments)} Compartments  |  '
            f'{len(reaction_df)} Active Reactions',
            fontsize=14, color='#58a6ff', fontweight='bold')

for fmt in ['png', 'svg', 'pdf']:
    filename = f"compartment_flux_map.{fmt}"
    plt.savefig(filename, facecolor='#0d1117', edgecolor='none')
    print(f"  ✓ Saved → {filename}")

plt.close()


# =============================================================================
# FINAL SUMMARY
# =============================================================================

print("\n" + "=" * 70)
print("COMPARTMENTALIZED PATHWAY MAP — COMPLETE")
print("=" * 70)
print(f"\n  Model: {MODEL_FILE}")
print(f"  Growth rate: {growth:.6f} mmol/gDW/h")
print(f"  Compartments analyzed: {len(compartments)}")
print(f"  Active reactions: {len(reaction_df)}")
print(f"  Transport reactions: {len(reaction_df[reaction_df['is_transport']])}")
print(f"  Active metabolites: {len(met_hub_df)}")
print("\n  Output files:")
print("    compartment_summary.csv              — flux summary per compartment")
print("    compartment_pathways.csv             — pathway distribution per compartment")
print("    transport_reactions_2.csv            — inter-compartment transport")
print("    compartment_metabolites.csv          — metabolite hubs per compartment")
print("    compartment_flux_map.png/svg/pdf     — visual pathway map")
print("    compartment_network.graphml          — full network (Cytoscape)")
print("\n  Next steps:")
print("    1. Open compartment_network.graphml in Cytoscape")
print("    3. Use compartment_flux_map.png for Analysis")
print("=" * 70)
```

```python
"""
=============================================================================
VFA/H+ Production Pathway Analysis Panel — FIXED VERSION
=============================================================================

FIXES IN THIS VERSION:
  1. Properly detects VFAs using ModelSEED IDs (cpd00029, cpd00211, etc.)
  2. Identifies exchange reactions (EX_ prefix) for secretion flux
  3. Tracks both cytosolic production AND extracellular secretion
  4. Includes succinate detection (cpd00036)
  5. Cross-references with biomass-coupled pathways

OUTPUT:
  - vfa_h2_panel.png/pdf          — Figure
  - vfa_h2_reactions.xlsx         — Excel workbook with all pathway data
  - vfa_h2_summary_table.csv      — Summary table
  - mechanistic_diagram.png       — Pathway linkage diagram

REQUIREMENTS
------------
  pip install cobra pandas numpy matplotlib openpyxl

USAGE
-----
  python vfa_h2_panel_fixed.py

# # =============================================================================
# # """

import cobra
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch, Circle, Rectangle
import warnings
warnings.filterwarnings("ignore")

# Configuration
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans']
plt.rcParams['font.size'] = 9
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

# File paths
MODEL_FILE = "Model"
BIOMASS_RXN = "bio1"
FLUX_CUTOFF = 1e-6

# VFA ModelSEED compound IDs
VFA_COMPOUNDS = {
    'Acetate': 'cpd00029',
    'Butyrate': 'cpd00211',
    'Propionate': 'cpd00141',
    'Formate': 'cpd00047',
    'Lactate': 'cpd00159',
    'Succinate': 'cpd00036',
    'Valerate': 'cpd00252',
    'Isobutyrate': 'cpd00221',
}

# H2 ModelSEED ID
H_COMPOUND = 'cpd00067'

# Pathway keywords
PATHWAY_KEYWORDS = {
    'Acetate Production': ['acetate kinase', 'phosphotransacetylase', 'pta', 'ack',
                          'acetyl-coa synthetase', 'acetyl phosphate', 'acetate--coa ligase'],
    'Butyrate Production': ['butyryl', 'butyrate kinase', 'phosphotransbutyrylase',
                           'crotonyl', 'beta-hydroxybutyryl', 'butanoate'],
    'Propionate Production': ['propionyl', 'methylmalonyl', 'propionate kinase',
                             'propanoate'],
    'Formate Production': ['pyruvate formate lyase', 'formate dehydrogenase',
                          'pfl', 'fdh', 'formate--tetrahydrofolate ligase'],
    'Lactate Production': ['lactate dehydrogenase', 'ldh', 'l-lactate', 'd-lactate'],
    'Succinate Production': ['succinate dehydrogenase', 'succinyl-coa', 'fumarate reductase',
                            'succinate--coa ligase', 'alpha-ketoglutarate'],
    'H2 Production': ['hydrogenase', 'ferredoxin', 'pyruvate:ferredoxin oxidoreductase',
                     'nadh:ferredoxin oxidoreductase'],
    'TCA Cycle': ['citrate', 'isocitrate', 'oxaloacetate', 'malate', 'fumarate',
                  'alpha-ketoglutarate'],
    'Pyruvate Metabolism': ['pyruvate', 'pyruvate decarboxylase', 'pyruvate kinase',
                           'pyruvate carboxylase'],
}


# =============================================================================
# 1. LOAD MODEL AND RUN FBA
# =============================================================================

print("=" * 70)
print("VFA/H+ PRODUCTION PATHWAY ANALYSIS — FIXED VERSION")
print("=" * 70)
print(f"\nLoading model: {MODEL_FILE}")

try:
    model = cobra.io.read_sbml_model(MODEL_FILE)
    print(f"  ✓ Model loaded: {len(model.reactions)} reactions, "
          f"{len(model.metabolites)} metabolites")
except Exception as e:
    print(f"[ERROR] Failed to load model: {e}")
    exit(1)

# Run FBA
print("\nRunning FBA...")
try:
    solution = model.optimize()
    if solution.status != "optimal":
        print(f"[ERROR] FBA failed with status: {solution.status}")
        exit(1)
    
    flux_series = solution.fluxes
    growth = solution.objective_value
    print(f"  ✓ FBA complete: growth = {growth:.6f} mmol/gDW/h")
    print(f"  ✓ Active reactions: {(flux_series.abs() > FLUX_CUTOFF).sum()}")
except Exception as e:
    print(f"[ERROR] FBA optimization failed: {e}")
    exit(1)


# =============================================================================
# 2. IDENTIFY VFA METABOLITES IN THE MODEL
# =============================================================================

print("\n" + "=" * 70)
print("Identifying VFA metabolites in the model...")

vfa_metabolites_found = {}

for vfa_name, cpd_id in VFA_COMPOUNDS.items():
    # Search for this compound ID in metabolite IDs
    found_mets = [met for met in model.metabolites if cpd_id in met.id]
    
    if found_mets:
        vfa_metabolites_found[vfa_name] = found_mets
        print(f"  ✓ {vfa_name:<15} ({cpd_id}): {len(found_mets)} metabolites")
        for met in found_mets:
            print(f"      {met.id:<25} {met.name or '(no name)':<40} [{met.compartment}]")
    else:
        print(f"  ✗ {vfa_name:<15} (not found in model)")

# H2 metabolites
h2_metabolites = [met for met in model.metabolites if H_COMPOUND in met.id]
print(f"\n  ✓ H+ metabolites: {len(h2_metabolites)}")
for met in h2_metabolites:
    print(f"      {met.id:<25} {met.name or '(no name)':<40} [{met.compartment}]")


# =============================================================================
# 3. ANALYZE VFA EXCHANGE REACTIONS (SECRETION)
# =============================================================================

print("\n" + "=" * 70)
print("Analyzing VFA exchange reactions (secretion)...")

vfa_secretion = []

for vfa_name, mets in vfa_metabolites_found.items():
    for met in mets:
        # Check exchange reactions for this metabolite
        for rxn in met.reactions:
            # Exchange reactions typically have EX_ prefix or are boundary reactions
            is_exchange = met.id.endswith('_e0')
            
            if not is_exchange:
                continue
            
            flux = flux_series.get(rxn.id, 0.0)
            if abs(flux) < FLUX_CUTOFF:
                continue
            
            # Positive flux in exchange = secretion (export from system)
            # Check compartment and direction
            stoich = rxn.metabolites[met]
            
            vfa_secretion.append({
                'vfa_type': vfa_name,
                'metabolite_id': met.id,
                'metabolite_name': met.name or met.id,
                'compartment': met.compartment,
                'reaction_id': rxn.id,
                'reaction_name': rxn.name or rxn.id,
                'flux': round(flux, 6),
                'secretion_rate': abs(flux),
                'direction': 'export' if flux > 0 else 'import',
            })

vfa_secretion_df = pd.DataFrame(vfa_secretion)

if len(vfa_secretion_df) > 0:
    print(f"\n  ✓ Found {len(vfa_secretion_df)} VFA exchange reactions")
    
    secretion_summary = (
        vfa_secretion_df[vfa_secretion_df['direction'] == 'export']
        .groupby('vfa_type')
        .agg(
            n_reactions=('reaction_id', 'count'),
            total_secretion=('secretion_rate', 'sum'),
        )
        .sort_values('total_secretion', ascending=False)
    )
    
    print("\n  VFA secretion rates (export):")
    for vfa_type, row in secretion_summary.iterrows():
        print(f"    {vfa_type:<15}: {row['total_secretion']:>10.6f} mmol/gDW/h")
else:
    print("  [WARNING] No VFA exchange reactions detected")
    secretion_summary = pd.DataFrame()


# =============================================================================
# 4. IDENTIFY VFA PRODUCTION REACTIONS (CYTOSOLIC)
# =============================================================================

print("\n" + "=" * 70)
print("Identifying cytosolic VFA production reactions...")

vfa_production = []

for vfa_name, mets in vfa_metabolites_found.items():
    for met in mets:
        # Focus on cytosolic metabolites
        if met.compartment not in ['c', 'c0']: #(Change the Compartments According to GEM/GCM)
            continue
        
        for rxn in met.reactions:
            flux = flux_series.get(rxn.id, 0.0)
            if abs(flux) < FLUX_CUTOFF:
                continue
            
            # Skip exchange reactions
            if rxn.id.startswith('EX_') or met.id.endswith('_e0') or rxn.boundary:
                continue
            
            # Check if this metabolite is produced
            stoich = rxn.metabolites[met]
            net_stoich = stoich * (1 if flux > 0 else -1)
            
            if net_stoich <= 0:  # Not a product
                continue
            
            # Classify pathway
            pathway = "Other"
            rxn_text = (rxn.id + " " + (rxn.name or "")).lower()
            for pw_name, pw_keywords in PATHWAY_KEYWORDS.items():
                if any(kw in rxn_text for kw in pw_keywords):
                    pathway = pw_name
                    break
            
            vfa_production.append({
                'vfa_type': vfa_name,
                'reaction_id': rxn.id,
                'reaction_name': rxn.name or rxn.id,
                'flux': round(flux, 6),
                'abs_flux': abs(flux),
                'pathway': pathway,
                'equation': rxn.reaction,
                'subsystem': rxn.subsystem or "Unknown",
            })

vfa_production_df = pd.DataFrame(vfa_production)

if len(vfa_production_df) > 0:
    print(f"  ✓ Found {len(vfa_production_df)} VFA production reactions")
    
    production_summary = (
        vfa_production_df.groupby('vfa_type')
        .agg(
            n_reactions=('reaction_id', 'count'),
            total_production=('abs_flux', 'sum'),
        )
        .sort_values('total_production', ascending=False)
    )
    
    print("\n  Cytosolic VFA production:")
    for vfa_type, row in production_summary.iterrows():
        print(f"    {vfa_type:<15}: {row['n_reactions']:>2} reactions, "
              f"flux = {row['total_production']:>10.6f} mmol/gDW/h")


# =============================================================================
# 5A. IDENTIFY H+ PRODUCTION REACTIONS (CYTOSOLIC)
# =============================================================================
 
print("\n" + "=" * 70)
print("Identifying H+ production reactions (cytosolic)...")
 
h2_production = []
 
for met in h2_metabolites:
    # Focus on cytosolic H2 first
    if met.compartment not in ['c', 'c0']: #(Change the Compartments According to GEM/GCM)
        continue
    
    for rxn in met.reactions:
        flux = flux_series.get(rxn.id, 0.0)
        if abs(flux) < FLUX_CUTOFF:
            continue
        
        # Skip exchange reactions
        if rxn.id.startswith('EX_') or met.id.endswith('_e0') or rxn.boundary:
            continue
        
        # Check if H+ is produced
        stoich = rxn.metabolites[met]
        net_stoich = stoich * (1 if flux > 0 else -1)
        
        if net_stoich <= 0:  # Not production
            continue
        
        # Classify pathway
        pathway = "H+ Production"
        rxn_text = (rxn.id + " " + (rxn.name or "")).lower()
        for pw_name, pw_keywords in PATHWAY_KEYWORDS.items():
            if any(kw in rxn_text for kw in pw_keywords):
                pathway = pw_name
                break
        
        h2_production.append({
            'metabolite_id': met.id,
            'reaction_id': rxn.id,
            'reaction_name': rxn.name or rxn.id,
            'flux': round(flux, 6),
            'abs_flux': abs(flux),
            'pathway': pathway,
            'equation': rxn.reaction,
            'subsystem': rxn.subsystem or "Unknown",
        })
 
h2_production_df = pd.DataFrame(h2_production)
 
if len(h2_production_df) > 0:
    print(f"  ✓ Found {len(h2_production_df)} H+ production reactions (cytosolic)")
    total_h2_production = h2_production_df['abs_flux'].sum()
    print(f"  ✓ Total cytosolic H+ production: {total_h2_production:.6f} mmol/gDW/h")
    
    print("\n  Top H+-producing reactions:")
    for _, row in h2_production_df.nlargest(5, 'abs_flux').iterrows():
        print(f"    {row['reaction_id']:<20} {row['flux']:>10.6f}  [{row['pathway']}]")
else:
    print("  [WARNING] No cytosolic H+ production reactions detected")
    total_h2_production = 0

# =============================================================================
# 5B. IDENTIFY H+ SECRETION (EXCHANGE REACTIONS)
# =============================================================================
 
print("\n" + "=" * 70)
print("Analyzing H+ secretion (exchange reactions)...")
 
h2_secretion = []
 
for met in h2_metabolites:
    # Check exchange reactions for this H+ metabolite
    for rxn in met.reactions:
        # Exchange reactions
        is_exchange = met.id.endswith('_e0')
        
        if not is_exchange:
            continue
        
        flux = flux_series.get(rxn.id, 0.0)
        if abs(flux) < FLUX_CUTOFF:
            continue
        
        stoich = rxn.metabolites[met]
        
        h2_secretion.append({
            'metabolite_id': met.id,
            'metabolite_name': met.name or met.id,
            'compartment': met.compartment,
            'reaction_id': rxn.id,
            'reaction_name': rxn.name or rxn.id,
            'flux': round(flux, 6),
            'secretion_rate': abs(flux),
            'direction': 'export' if flux > 0 else 'import',
        })
 
h2_secretion_df = pd.DataFrame(h2_secretion)
 
if len(h2_secretion_df) > 0:
    print(f"  ✓ Found {len(h2_secretion_df)} H+ exchange reactions")
    
    h2_export = h2_secretion_df[h2_secretion_df['direction'] == 'export']
    total_h2_secretion = h2_export['secretion_rate'].sum()
    
    print(f"  ✓ Total H+ secretion (export): {total_h2_secretion:.6f} mmol/gDW/h")
    
    if len(h2_export) > 0:
        print("\n  H+ export reactions:")
        for _, row in h2_export.iterrows():
            print(f"    {row['reaction_id']:<20} {row['flux']:>10.6f} mmol/gDW/h")
else:
    print("  [INFO] No H+ exchange reactions detected")
    total_h2_secretion = 0
 
# Combined H+ metrics
total_h2 = total_h2_production  # Use production as primary metric


# =============================================================================
# 6. CREATE SUMMARY TABLE FOR MANUSCRIPT
# =============================================================================

print("\n" + "=" * 70)

summary_table = []

# Combine secretion and production data
for vfa_name in VFA_COMPOUNDS.keys():
    # Secretion rate
    secretion = 0
    if len(vfa_secretion_df) > 0:
        sec_data = vfa_secretion_df[
            (vfa_secretion_df['vfa_type'] == vfa_name) &
            (vfa_secretion_df['direction'] == 'export')
        ]
        secretion = sec_data['secretion_rate'].sum()
    
    # Production pathways
    production = 0
    n_pathways = 0
    pathways_list = []
    if len(vfa_production_df) > 0:
        prod_data = vfa_production_df[vfa_production_df['vfa_type'] == vfa_name]
        production = prod_data['abs_flux'].sum()
        pathways_list = prod_data['pathway'].unique().tolist()
        n_pathways = len(pathways_list)
    
    # Only include if either secretion or production > 0
    if secretion > FLUX_CUTOFF or production > FLUX_CUTOFF:
        summary_table.append({
            'Metabolite': vfa_name,
            'ModelSEED_ID': VFA_COMPOUNDS[vfa_name],
            'Secretion_Rate_mmol_gDW_h': round(secretion, 6),
            'Cytosolic_Production_mmol_gDW_h': round(production, 6),
            'N_Production_Pathways': n_pathways,
            'Key_Pathways': '; '.join(pathways_list[:3]) if pathways_list else 'None detected',
        })

# Add H+
if total_h2 > FLUX_CUTOFF:
    h2_pathways = h2_production_df['pathway'].unique().tolist()
    summary_table.append({
        'Metabolite': 'Hydrogen (H+)',
        'ModelSEED_ID': H_COMPOUND,
        'Secretion_Rate_mmol_gDW_h': 0.0,  # H+ usually detected as production, not exchange
        'Cytosolic_Production_mmol_gDW_h': round(total_h2, 6),
        'N_Production_Pathways': len(h2_pathways),
        'Key_Pathways': '; '.join(h2_pathways[:3]) if h2_pathways else 'H+ Production',
    })

summary_df = pd.DataFrame(summary_table)

if len(summary_df) > 0:
    summary_df = summary_df.sort_values('Secretion_Rate_mmol_gDW_h', ascending=False)
    summary_df.to_csv("vfa_h2_summary_table.csv", index=False)
    print("  ✓ Saved → vfa_h2_summary_table.csv")
    
    print("\n" + "=" * 70)
    print("VFA/H+ PRODUCTION SUMMARY TABLE")
    print("=" * 70)
    print(summary_df.to_string(index=False))
else:
    print("  [WARNING] No VFA/H+ production detected in model")


# =============================================================================
# 7. EXPORT TO EXCEL
# =============================================================================

print("\n" + "=" * 70)
print("Exporting to Excel...")
 
try:
    with pd.ExcelWriter("vfa_h2_reactions.xlsx", engine='openpyxl') as writer:
        if len(summary_df) > 0:
            summary_df.to_excel(writer, sheet_name='Summary_Table', index=False)
            print("  ✓ Sheet: Summary_Table")
        
        if len(vfa_secretion_df) > 0:
            vfa_secretion_df.to_excel(writer, sheet_name='VFA_Secretion', index=False)
            print("  ✓ Sheet: VFA_Secretion")
        
        if len(vfa_production_df) > 0:
            vfa_production_df.to_excel(writer, sheet_name='VFA_Production', index=False)
            print("  ✓ Sheet: VFA_Production")
        
        if len(h2_secretion_df) > 0:
            h2_secretion_df.to_excel(writer, sheet_name='H+_Secretion', index=False)
            print("  ✓ Sheet: H+_Secretion")
        
        if len(h2_production_df) > 0:
            h2_production_df.to_excel(writer, sheet_name='H+_Production', index=False)
            print("  ✓ Sheet: H+_Production")
        
        # Auto-adjust column widths
        for sheet_name in writer.sheets:
            worksheet = writer.sheets[sheet_name]
            for column in worksheet.columns:
                max_length = 0
                column_letter = column[0].column_letter
                for cell in column:
                    try:
                        if len(str(cell.value)) > max_length:
                            max_length = len(str(cell.value))
                    except:
                        pass
                adjusted_width = min(max_length + 2, 60)
                worksheet.column_dimensions[column_letter].width = adjusted_width
            worksheet.freeze_panes = 'A2'
    
    print("  ✓ Saved → vfa_h2_reactions.xlsx")
 
except Exception as e:
    print(f"  [ERROR] Failed to create Excel file: {e}")

# =============================================================================
# 8. CREATE PUBLICATION-READY FIGURE
# =============================================================================

print("\n" + "=" * 70)

if len(summary_df) == 0:
    print("  [SKIP] No data to visualize")
else:
    fig = plt.figure(figsize=(16, 10))
    fig.patch.set_facecolor('white')
    
    gs = fig.add_gridspec(2, 2, hspace=0.35, wspace=0.35,
                         left=0.08, right=0.95, top=0.92, bottom=0.08)
    
    # ── Panel A: VFA Secretion Rates ──
    ax1 = fig.add_subplot(gs[0, 0])
    
    vfa_only = summary_df[summary_df['Metabolite'] != 'Hydrogen (H+)'].copy()
    if len(vfa_only) > 0:
        y_pos = np.arange(len(vfa_only))
        colors = plt.cm.Set3(np.linspace(0, 1, len(vfa_only)))
        
        ax1.barh(y_pos, vfa_only['Secretion_Rate_mmol_gDW_h'], 
                color=colors, edgecolor='black', linewidth=1.2)
        
        ax1.set_yticks(y_pos)
        ax1.set_yticklabels(vfa_only['Metabolite'], fontsize=10)
        ax1.set_xlabel('Secretion Rate (mmol/gDW/h)', fontsize=11, fontweight='bold')
        ax1.set_title('A. VFA Secretion Rates', fontsize=12, fontweight='bold', loc='left')
        ax1.spines['top'].set_visible(False)
        ax1.spines['right'].set_visible(False)
        ax1.grid(axis='x', alpha=0.3, linestyle='--')
        
        # Add values
        for i, (_, row) in enumerate(vfa_only.iterrows()):
            if row['Secretion_Rate_mmol_gDW_h'] > 0:
                ax1.text(row['Secretion_Rate_mmol_gDW_h'] * 1.02, i,
                        f"{row['Secretion_Rate_mmol_gDW_h']:.4f}",
                        va='center', fontsize=8, fontweight='bold')
    else:
        ax1.text(0.5, 0.5, 'No VFA secretion detected', 
                ha='center', va='center', fontsize=11)
        ax1.set_xlim(0, 1)
        ax1.set_ylim(0, 1)
        ax1.axis('off')
    
    # ── Panel B: Cytosolic Production vs Secretion ──
    ax2 = fig.add_subplot(gs[0, 1])
    
    if len(vfa_only) > 0:
        x = vfa_only['Cytosolic_Production_mmol_gDW_h']
        y = vfa_only['Secretion_Rate_mmol_gDW_h']
        colors_scatter = plt.cm.Set3(np.linspace(0, 1, len(vfa_only)))
        
        ax2.scatter(x, y, s=200, c=colors_scatter, edgecolors='none', 
                   linewidths=1.5, alpha=0.8, zorder=3)
        
        # Labels
        for i, (_, row) in enumerate(vfa_only.iterrows()):
            ax2.annotate(row['Metabolite'], 
                        (row['Cytosolic_Production_mmol_gDW_h'],
                         row['Secretion_Rate_mmol_gDW_h']),
                        xytext=(5, 5), textcoords='offset points',
                        fontsize=8, fontweight='bold')
        
        # Diagonal reference line
        max_val = max(x.max(), y.max()) * 1.1
        ax2.plot([0, max_val], [0, max_val], 'k--', alpha=0.3, linewidth=1, 
                label='Production = Secretion')
        
        ax2.set_xlabel('Cytosolic Production (mmol/gDW/h)', fontsize=11, fontweight='bold')
        ax2.set_ylabel('Secretion Rate (mmol/gDW/h)', fontsize=11, fontweight='bold')
        ax2.set_title('B. Production vs Secretion', fontsize=12, fontweight='bold', loc='left')
        ax2.spines['top'].set_visible(False)
        ax2.spines['right'].set_visible(False)
        ax2.grid(alpha=0.3, linestyle='--')
        ax2.legend(fontsize=9, loc='upper left')
    else:
        ax2.text(0.5, 0.5, 'No data', ha='center', va='center')
        ax2.axis('off')
    
    # ── Panel C: Pathway Contribution ──
    ax3 = fig.add_subplot(gs[1, 0])
    
    if len(vfa_production_df) > 0:
        pathway_contrib = (
            vfa_production_df.groupby(['vfa_type', 'pathway'])
            .agg(flux=('abs_flux', 'sum'))
            .reset_index()
        )
        
        # Pivot for stacked bar
        pivot = pathway_contrib.pivot(index='vfa_type', columns='pathway', values='flux')
        pivot = pivot.fillna(0)
        
        pivot.plot(kind='bar', stacked=True, ax=ax3, 
                  colormap='tab20', edgecolor='black', linewidth=0.8)
        
        ax3.set_xlabel('VFA Type', fontsize=11, fontweight='bold')
        ax3.set_ylabel('Total Production Flux (mmol/gDW/h)', fontsize=11, fontweight='bold')
        ax3.set_title('C. Pathway Contribution to VFA Production', 
                     fontsize=12, fontweight='bold', loc='left')
        ax3.legend(fontsize=8, loc='upper left', ncol=2)
        ax3.spines['top'].set_visible(False)
        ax3.spines['right'].set_visible(False)
        ax3.grid(axis='y', alpha=0.3, linestyle='--')
        plt.setp(ax3.xaxis.get_majorticklabels(), rotation=0, ha='right')
    else:
        ax3.text(0.5, 0.5, 'No pathway data', ha='center', va='center')
        ax3.axis('off')
    
    # ── Panel D: H+ Production ──
    ax4 = fig.add_subplot(gs[1, 1])
    
    if len(h2_production_df) > 0:
        h2_by_pathway = (
            h2_production_df.groupby('pathway')
            .agg(flux=('abs_flux', 'sum'))
            .sort_values('flux', ascending=True)
        )
        
        y_pos = np.arange(len(h2_by_pathway))
        colors = plt.cm.Blues(np.linspace(0.4, 0.9, len(h2_by_pathway)))
        
        ax4.barh(y_pos, h2_by_pathway['flux'], color=colors, 
                edgecolor='black', linewidth=1.2)
        
        ax4.set_yticks(y_pos)
        ax4.set_yticklabels(h2_by_pathway.index, fontsize=10)
        ax4.set_xlabel('H+ Production Flux (mmol/gDW/h)', fontsize=11, fontweight='bold')
        ax4.set_title(f'D. H+ Production by Pathway (Total: {total_h2:.4f} mmol/gDW/h)',
                     fontsize=12, fontweight='bold', loc='left')
        ax4.spines['top'].set_visible(False)
        ax4.spines['right'].set_visible(False)
        ax4.grid(axis='x', alpha=0.3, linestyle='--')
        
        # Add values
        for i, (pathway, row) in enumerate(h2_by_pathway.iterrows()):
            ax4.text(row['flux'] * 1.02, i, f"{row['flux']:.4f}",
                    va='center', fontsize=8, fontweight='bold')
    else:
        ax4.text(0.5, 0.5, 'No H+ production detected', 
                ha='center', va='center', fontsize=11)
        ax4.axis('off')
    
    # Main title
    fig.suptitle(f'VFA/H+ Production Panel — Model\n'
                f'Growth: {growth:.6f} mmol/gDW/h',
                fontsize=14, fontweight='bold')
    
    for fmt in ['png', 'pdf', 'svg']:
        filename = f"vfa_h2_panel.{fmt}"
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"  ✓ Saved → {filename}")
    
    plt.close()

# =============================================================================
# FINAL SUMMARY
# =============================================================================

print("\n" + "=" * 70)
print("VFA/H+ ANALYSIS COMPLETE")
print("=" * 70)
print(f"\nModel: {MODEL_FILE}")
print(f"Growth rate: {growth:.6f} mmol/gDW/h")
print(f"\nVFA/H+ detected:")
if len(summary_df) > 0:
    for _, row in summary_df.iterrows():
        print(f"  {row['Metabolite']:<20}: "
              f"Secretion = {row['Secretion_Rate_mmol_gDW_h']:.6f}, "
              f"Production = {row['Cytosolic_Production_mmol_gDW_h']:.6f} mmol/gDW/h")
else:
    print("  [NONE DETECTED]")

print("\nOutput files:")
print("  vfa_h2_summary_table.csv   — Summary table")
print("  vfa_h2_reactions.xlsx      — Detailed reaction data (Excel)")
print("  vfa_h2_panel.png/pdf/svg       — VFA/H+ Panel figure")
print("  Use vfa_h2_summary_table.csv as Table for Analysis")
print("  Use vfa_h2_panel.png as Figure showing mechanistic pathways")
print("=" * 70)
```

```python
# Calculation of Pearson's r Correlation

import numpy as np
from scipy.stats import pearsonr
from scipy.stats import norm

# Example predicted and observed values
predicted = np.array([0.198, 0.710, 0.404, 0.210, 0.150, 0.210])
observed  = np.array([0.176, 0.689, 0.363, 0.187, 0.136, 0.189])

# Pearson correlation
r, p = pearsonr(predicted, observed)

# Fisher Z transformation
n = len(predicted)
z = np.arctanh(r)
se = 1 / np.sqrt(n - 3)

# 95% CI
z_critical = norm.ppf(0.975)

z_lower = z - z_critical * se
z_upper = z + z_critical * se

# Convert back to r
r_lower = np.tanh(z_lower)
r_upper = np.tanh(z_upper)

print(f"Pearson r: {r:.4f}")
print(f"95% CI for r: [{r_lower:.4f}, {r_upper:.4f}]")
```
